```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
import find_nearest_venues as fnv
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
train_events_sparce = pd.read_csv('train_events_sparce.csv')
test_events_sparce = pd.read_csv('test_events_sparce.csv')
train_venues = pd.read_csv('train_venues.csv')
artists = pd.read_csv('artists.csv')
test_events = pd.read_csv('test_events.csv')
master_venues = pd.read_csv('master_venues.csv')
```

```python
def find_nearest_neighbor_by_genre(artist, venues):
    if artist.genre.isna().any(axis=0):
        return None
    cv = CountVectorizer(stop_words='english', max_features=100)
    venues_with_genres = venues[~venues.venue_genre.isnull()]
    venue_vectors = cv.fit_transform(venues_with_genres.venue_genre)
    artist_vector = cv.transform(artist.genre)
    artist_neighbors = cosine_similarity(venue_vectors, artist_vector, dense_output=True)
    a = list((np.argsort(-artist_neighbors, axis = 0)[0:20]).reshape(1,-1)[0])
    neighbors = venues_with_genres.iloc[a,:]
    return list(neighbors.venue_id)
```

```python
def find_nearest_neighbor_by_bio(artist, venues):
    if artist.bio.isna().any(axis=0):
        return None
    cv = CountVectorizer(stop_words='english', max_features=800)
    venues_with_bios = venues[~venues.venue_bio.isnull()]
    venue_vectors = cv.fit_transform(venues_with_bios.venue_bio)
    artist_vector = cv.transform(artist.bio)
    artist_neighbors = cosine_similarity(venue_vectors, artist_vector, dense_output=True)
    a = list((np.argsort(-artist_neighbors, axis = 0)[0:20]).reshape(1,-1)[0])
    neighbors = venues_with_bios.iloc[a,:]
    return list(neighbors.venue_id)
```

```python
def get_some_events(n):
    samples = test_events.sample(n)
    return samples
```

```python
samples = get_some_events(90)
```

```python
for i, sample in samples.iterrows():
    artist_id = sample.artist_id
    artist = artists[artists.artists_id == artist_id]
    if artist.genre.isna().any(axis=0):
        continue
    venue_id = sample['venue_id']
    if venue_id not in train_venues.venue_id:
        continue
    venue = train_venues[train_venues['venue_id'] == venue_id]
    if (len(venue.latitude)) == 0 :
        continue
    venues = fnv.get_nearest_venues(venue_id, train_venues, master_venues)
    #print (venue_id in list(venues.venue_id), venue_id in train_venues.venue_id)
    recomendations = find_nearest_neighbor_by_genre(artist, venues)
    if venue_id in recomendations:
        print ('secess')
    else:
        print ('failure')
    
    
```
